Программу **Python** для создания и отображения **кривой Мандельброта** с использованием графики черепахи.

**Множество Мандельброта** — это известный фрактал, демонстрирующий замысловатые и сложные узоры.

Вот программа Python для создания и отображения кривой Мандельброта с использованием графики черепахи:

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors

# Настройте параметры
width, height = 800, 800
x_min, x_max = -2.5, 1.0
y_min, y_max = -1.5, 1.5
max_iter = 100

# Создайте массив комплексных чисел, соответствующих каждому пикселю
x = np.linspace(x_min, x_max, width)
y = np.linspace(y_min, y_max, height)
X, Y = np.meshgrid(x, y)
c = X + 1j * Y

# Создайте пустой массив для хранения набора Мандельброта
mandelbrot = np.zeros((height, width))

# Выполните итерацию для каждого пикселя
z = np.zeros_like(c)
for i in range(max_iter):
    mask = np.abs(z) <= 2
    z[mask] = z[mask]**2 + c[mask]
    mandelbrot += mask.astype(int)

# Нормируйте массив множества Мандельброта к значениям между 0 и 1
mandelbrot /= max_iter

# Определите пользовательскую палитру для окрашивания набора Мандельброта градиентом цветов.
color_map = colors.LinearSegmentedColormap.from_list('mandelbrot_colors',
                                                     ['#000000', '#0C056D', '#590D82', '#B00A6B', '#E82F2C', '#FF6E00',
                                                      '#FFD500', '#AAFF32', '#05FA15', '#B00A6B'], N=max_iter)

# Постройте и сохраните множество Мандельброта с цветным интерьером
plt.imshow(mandelbrot, cmap=color_map, extent=[x_min, x_max, y_min, y_max])
plt.title("Кривая Мандельброта")
plt.axis('off')
plt.savefig("mandelbrot.png", dpi=300, bbox_inches='tight')
plt.show()
```

    3. **matplotlib.colors** для создания пользовательских цветовых карт.

    4. **max_iter **определяет максимальное количество итераций для каждого пикселя.

    2. Итерировать **max_iter** раз:
     Создайте маску, чтобы определить, какие элементы **z** имеют абсолютное значение меньше или равное **2**.
