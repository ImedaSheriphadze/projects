Создание онлайн-калькулятора на основе нейронной сети, способного решать произвольные, очень сложные математические уравнения, — очень амбициозная задача, выходящая за рамки возможностей простой программы. Решение произвольных уравнений обычно включает в себя символьную математику и продвинутые методы, такие как символическое интегрирование и дифференцирование, с которыми одни лишь нейронные сети не могут эффективно справиться. Кроме того, определить процент правильности и точности такой системы чрезвычайно сложно.

Однако я могу предоставить вам упрощенный пример того, как можно использовать символьную математику и нейронную сеть для аппроксимации решений уравнений определенной формы. В этом примере мы аппроксимируем решения квадратных уравнений:

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Определим квадратное уравнение: ax^2 + bx + c = 0 (Здесь вы можете найти любое уравнение, которое захотите.)
# Не забудьте указать соответствующие переменные
a = 1.0
b = -3.0
c = 2.0

# Генерация синтетических данных
X = np.linspace(-10, 10, 1000)
Y = a * X**2 + b * X + c

# Создайте простую модель нейронной сети
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])

# Скомпилируйте модель
model.compile(optimizer='adam', loss='mean_squared_error')

# Обучите модель аппроксимировать уравнение
model.fit(X, Y, epochs=100, verbose=0)

# Создайте график соответствующей функции
X_plot = np.linspace(-10, 10, 1000)
Y_predicted = model.predict(X_plot)

# Постройте истинные данные
plt.plot(X, Y, label="Истинная функция")

# Plot the predicted data
plt.plot(X_plot, Y_predicted, label="Прогнозируемая функция", color='r')

plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Приближение нейронной сети для квадратного уравнения")
plt.show()
```

В этом упрощенном примере мы аппроксимируем решения квадратного уравнения с помощью нейронной сети. Обратите внимание, что это упрощенная задача, и для более сложных уравнений или общих уравнений потребуется гораздо более сложный подход и, возможно, сочетание символьной математики и методов машинного обучения. Процент правильности и точности для произвольных уравнений определить нелегко.

Решение произвольных математических уравнений остается сложной и сложной проблемой в информатике и математике и обычно требует специального программного обеспечения и инструментов.
