**Гипотеза Римана** — это нерешенная проблема в математике, с  2021 года не было разработано ни программы, ни алгоритма для доказательства или опровержения гипотезы Римана. Это остается одной из самых сложных и важных проблем теории чисел.

Хотя невозможно предоставить программу, решающую **гипотезу Римана**, я определенно могу написать программу для вычисления дзета-функции Римана и построить ее график с помощью **Python**. Эта программа может быть полезным инструментом для изучения поведения дзета-функции.

Вот пример программы, которая использует библиотеку **SciPy** для вычисления дзета-функции Римана и **Matplotlib** для построения графика:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import zeta

def plot_zeta_function(start, end, num_points):
    x = np.linspace(start, end, num_points)
    y = zeta(x)

    plt.plot(x, y)
    plt.xlabel('s')
    plt.ylabel('ζ(s)')
    plt.title('График дзета-функции Римана')
    plt.grid(True)
    plt.show()

# Пример использования
plot_zeta_function(0.1, 10, 100)
plot_zeta_function(0.1, 10, 1000)
plot_zeta_function(0.1, 10, 10000)
plot_zeta_function(0.1, 10, 100000)
```

В этой программе мы определяем функцию **plot_zeta_function**, которая принимает начальное и конечное значения **s**, а также количество точек, которые нужно сгенерировать между ними. Он использует **numpy.linspace** для создания диапазона значений **s**, а затем вычисляет соответствующие значения дзета-функции Римана, используя **scipy.special.zeta**. Наконец, он использует **matplotlib.pyplot** для построения графика дзета-функции.

Пожалуйста, имейте в виду, что эта программа только визуализирует дзета-функцию и не решает гипотезу Римана. Гипотеза Римана остается открытой проблемой, и поиск программы для ее решения потребует новаторского математического открытия.
