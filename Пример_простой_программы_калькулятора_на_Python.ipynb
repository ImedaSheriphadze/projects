Пример простой программы-калькулятора на **Python**, которая может выполнять основные арифметические операции, такие как сложение, вычитание, умножение и деление:

**Объяснение каждой команды:**

1. Определите функции для каждой арифметической операции - Эти функции принимают на вход два числа и выполняют над ними соответствующую арифметическую операцию, возвращая результат. Функция деления также проверяет деление на ноль и обрабатывает случай ошибки.

2. Получить пользовательский ввод - Эти строки предлагают пользователю ввести два числа, которые затем сохраняются в переменных **num1** и **num2** после преобразования их в числа с плавающей запятой с помощью функции **float()**.

3. Распечатайте доступные операции и получите выбор пользователя - Эти строки отображают доступные арифметические операции и предлагают пользователю ввести свой выбор

4. Выполнить выбранную операцию - В этих строках используются условные операторы (**if**, **elif**, **else**) для определения выбранной операции на основе ввода пользователя. Соответствующая функция вызывается с аргументами **num1** и **num2**, а результат сохраняется в переменной результата. Выбранный символ операции также сохраняется в переменной операции.

5. Распечатать результат - Этот условный оператор проверяет, не равны ли **None** переменные результата и операции. Если они не равны **None**, уравнение и результат печатаются с использованием интерполяции форматированной строки.

```python
def add(x, y):
    """Функция сложения двух чисел"""
    return x + y

def subtract(x, y):
    """Функция вычитания двух чисел"""
    return x - y

def multiply(x, y):
    """Функция умножения двух чисел"""
    return x * y
def divide(x, y):
    """Функция деления двух чисел"""
    if y != 0:
        return x / y
    else:
        print("Ошибка: Деление на ноль!")
        return None

# Get user input
num1 = float(input("Введите первое число: "))
num2 = float(input("Введите второе число: "))

print("Выберите операцию:")
print("1. Добавление")
print("2. вычитание")
print("3. Умножение")
print("4. Разделение")

choice = input("Введите свой выбор (1-4): ")

# Выполнить выбранную операцию
if choice == '1':
    result = add(num1, num2)
    operation = '+'
elif choice == '2':
    result = subtract(num1, num2)
    operation = '-'
elif choice == '3':
    result = multiply(num1, num2)
    operation = '*'
elif choice == '4':
    result = divide(num1, num2)
    operation = '/'
else:
    print("Неверный Ввод!")
    result = None
    operation = None

# Распечатать результат
if result is not None and operation is not None:
    print(f"{num1} {operation} {num2} = {result}")
```
